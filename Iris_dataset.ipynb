```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the built-in iris dataset from seaborn
df = sns.load_dataset('iris')

# 1. Check dataset shape (Rows and Columns)
print("Dataset Shape (Rows, Columns):", df.shape)
print("-" * 50)

# 2. Display column names
print("Dataset Columns:", df.columns.tolist())
print("-" * 50)

# 3. View the first 5 rows of data
print("First 5 Rows of the Dataset:")
df.head()
```

```python
# Analyze the relationship between sepal length and sepal width
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='Set1')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
# Examine the data distribution of petal length
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='petal_length', kde=True, color='purple', bins=20)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Count')
plt.show()
```

```python
# Detect spread and potential outliers for petal width across species
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='species', y='petal_width', palette='Pastel1')
plt.title('Petal Width Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
# ### Conclusion & Insights:
# 1. Scatter Plot: The Setosa species forms a completely distinct cluster.
# 2. Histogram: The petal length distribution shows a clear bimodal gap.
# 3. Box Plot: Virginica generally possesses the highest petal width values.
```
